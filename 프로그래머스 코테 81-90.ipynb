```python
#81. 캐릭터의 좌표
def solution(keyinput, board):
    answer = [0, 0]
    
    # 제한된 좌표 범위 계산
    max_x = board[0] // 2
    max_y = board[1] // 2
    
    # 일반적인 이동 처리
    for key in keyinput:
        if key == "up":
            answer[1] = min(answer[1] + 1, max_y)  # y값 증가, 최대 max_y 초과 금지
        elif key == "down":
            answer[1] = max(answer[1] - 1, -max_y)  # y값 감소, 최소 -max_y 초과 금지
        elif key == "left":
            answer[0] = max(answer[0] - 1, -max_x)  # x값 감소, 최소 -max_x 초과 금지
        elif key == "right":
            answer[0] = min(answer[0] + 1, max_x)  # x값 증가, 최대 max_x 초과 금지

    return answer
```

```python
#82. 외계어 사전
from itertools import permutations

def solution(spell, dic):
    result = ["".join(i) for i in permutations(spell)] 

    if any(word in dic for word in result): # any 사용 (한번이라도 있으면)
        return 1
    else:
        return 2
```

```python
#84. 직사각형 넓이 구하기
def solution(dots):
    answer = 0
    max_value = 0
    min_value = 0
    min_index = 0
    max_index = 0
    sum_list = []
    for i in dots:
        sum_list.append(sum(i))
    max_value = max(sum_list)
    max_index = sum_list.index(max_value)
    min_value = min(sum_list)
    min_index = sum_list.index(min_value)

    answer = (dots[max_index][0] - dots[min_index][0]) * (dots[max_index][1] - dots[min_index][1])
    return answer
```

```python
#85. 로그인 성공?
def solution(id_pw, db):
    answer = ''
    db_id = []
    db_pw = []
    
    for i in db:
        db_id.append(i[0])
        db_pw.append(i[1])  

    if id_pw[0] not in db_id: 
        return 'fail'
    
    index = db_id.index(id_pw[0]) # 로그인의 인덱스와 비밀번호의 인덱스가 같아야 함
    
    if id_pw[1] == db_pw[index]:  
        return 'login'
    else:
        return 'wrong pw'
```

```python
#86. 치킨 쿠폰
def solution(chicken):
    answer = 0
    
    while chicken >= 10:
        quo = chicken // 10  
        res = chicken % 10   
        
        answer += quo  
        chicken = quo + res  
        
    return answer
```

```python
#87. 등수 매기기
def solution(score):
    score_average = [(i[0] + i[1]) / 2 for i in score]

    sorted_score = sorted(score_average, reverse=True)

    ranks = []
    for avg in score_average:
        rank = sorted_score.index(avg) + 1  
        ranks.append(rank)

    return ranks
```

```python
#88. 저주의 숫자 3
def solution(n):
    num_list = list(range(1, 201))  

    multiply3 = [i for i in range(1, 201) if '3' in str(i)]

    num_list_3 = [j for j in range(1, 201) if j % 3 == 0]

    town3 = set(num_list_3 + multiply3)

    num_list = [num for num in num_list if num not in town3]
    
    return num_list[n - 1]
```

```python
#89. 유한소수 판별하기
def solution(a, b):
    divisor = gcd(a, b)
    b //= divisor  

    while b % 2 == 0:
        b //= 2
    while b % 5 == 0:
        b //= 5

    if b == 1:
        return 1 
    else:
        return 2  

def gcd(m, n):
    while n != 0:
        m, n = n, m % n
    return m
```
